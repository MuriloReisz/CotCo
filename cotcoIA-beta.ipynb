```python
import pandas as pd
import re
import warnings
import random
```

```python
warnings.filterwarnings("ignore")
```

```python
file_path = '/content/electronics_product.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
df.head()
```

```python
# Criando a coluna de tempo de entrega e fazendo gerar numeros entre 1 a 12

df['Tempo_de_Entrega'] = [random.randint(1, 12) for _ in range(len(df))]

# Remover caracteres não numéricos da coluna 'actual_price'
df['actual_price'] = df['actual_price'].str.replace(r'[^\d.]', '', regex=True)

# Converter para float
df['actual_price'] = df['actual_price'].astype(float)

# Multiplicar os valores por 0.064 para converter de ₹ para R$
df['actual_price'] = df['actual_price'] * 0.064

# Formatar os valores para o formato X.XXX,XX
df['actual_price'] = df['actual_price'].apply(lambda x: '{:,.2f}'.format(x).replace('.', ',').replace(',', '.', 1))

# Exibir as primeiras linhas do DataFrame com os valores convertidos para reais
print(df.head())
```

Caso custo beneficio

```python
import pandas as pd
from IPython.display import display, Image


product_name = input()
produtos_filtrados = df[df['name'].str.contains(product_name, case=False)]

if not produtos_filtrados.empty:
    # Remover pontos dos milhares e substituir vírgula por ponto para transformar em float
    produtos_filtrados['actual_price'] = produtos_filtrados['actual_price'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.').astype(float)
    # Ordenar os produtos pelo preço
    produtos_filtrados = produtos_filtrados.sort_values(by='actual_price')

    # Selecionar os três produtos com maior custo-benefício
    top_produtos = produtos_filtrados.head(3)

    for index, produto in top_produtos.iterrows():
        print(f"Produto: {produto['name']}\nPreço: R$ {produto['actual_price']:.2f}")
        display(Image(url=produto['image']))
        print("\n")
else:
    print("Nenhum produto encontrado com este nome no arquivo anexado.")
```

Caso melhor avaliação

```python
product_name = input()
produtos_filtrados = df[df['name'].str.contains(product_name, case=False)]

if not produtos_filtrados.empty:
    # Ordenar os produtos pelo ratings
    produtos_filtrados = produtos_filtrados.sort_values(by='ratings', ascending=False)

    # Selecionar os três produtos com maior ratting
    top_produtos = produtos_filtrados.head(3)

    for index, produto in top_produtos.iterrows():
        print(f"Produto: {produto['name']}\nRatings: {produto['ratings']}")
        display(Image(url=produto['image']))
        print("\n")
else:
    print("Nenhum produto encontrado com este nome no arquivo anexado.")
```

# **Caso de melhor tempo de entrega**

```python
product_name = input()
produtos_filtrados = df[df['name'].str.contains(product_name, case=False)]

if not produtos_filtrados.empty:
    # Ordenar os produtos pelo Tempo_de_Entrega
    produtos_filtrados = produtos_filtrados.sort_values(by='Tempo_de_Entrega', ascending=False)

    # Selecionar os três produtos com maior Tempo_de_Entrega
    top_produtos = produtos_filtrados.head(3)

    for index, produto in top_produtos.iterrows():
        print(f"Produto: {produto['name']}\nTempo_de_Entrega: {produto['Tempo_de_Entrega']}")
        display(Image(url=produto['image']))
        print("\n")
else:
    print("Nenhum produto encontrado com este nome no arquivo anexado.")
```
